```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

출처: https://machinelearningmastery.com/tutorial-first-neural-network-python-keras/

```python
import os
```

```python
tf.config.list_physical_devices('GPU')
```

# 1. Load Data

```python
dataset = loadtxt(os.path.join('pima-indians-diabetes.csv'), delimiter=',')
dataset[0]
```

```python
x = dataset[:, 0:8]
y = dataset[:, 8]
```

# 2. Define Keras Model

```python
model = Sequential()
# 한 행에 8개의 데이터 입력, 12개 노드로 구성
model.add(Dense(12, input_shape=(8,), activation='relu'))
# 두번째 8개 노드로 구성
model.add(Dense(8, activation='relu'))
# 출력은 sigmoid로 구성
model.add(Dense(1, activation='sigmoid'))
```

# 3. Compile Keras Model

```python
# 이진 분류이기 때문에 'binary_crossentropy' 사용
# 분류 이기 때문에 'accuracy' 지정
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 4. Fit Keras Model

```python
model.fit(x, y, epochs=200, batch_size=10, verbose=True)
```

# 5. Evaluate Keras Model

```python
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % accuracy)
```

# 6. Make Prediction

```python
predictions = model.predict(x)
```
